## Exercise 2 — Add / update / remove

## Exercise 3 — Loop a dictionary ⭐
`marks` dict di gayi hai.
1. Sirf subjects (keys) loop karke print karo
2. Phir `.items()` se `subject -> score` print karo

```python
marks = {"Math": 90, "Physics": 75, "Chemistry": 88, "Biology": 95}

# TODO: loop keys only
for subject in marks:
    print(subject)

# TODO: loop with .items() -> "subject -> score"
for subject, mark in marks.items():
    print(subject, ":" , mark)
```

## Exercise 4 — Sets & duplicates
1. Is list se duplicates hatao using `set()`
2. Result print karo
3. Loop chala kar unique values print karo

```python
votes = ["PTI", "PMLN", "PTI", "PPP", "PMLN", "PTI"]

# TODO: make a set to remove duplicates
votes = set(votes)
print(votes)

# TODO: loop over the unique values and print each
for vote in votes:
    print(vote)
```

## Exercise 5 — Set operations
`group_a` aur `group_b` students ke sets hain.
1. Union print karo (sab students)
2. Intersection print karo (dono mein common)

```python
group_a = {"Ali", "Sara", "Bilal", "Hina"}
group_b = {"Bilal", "Hina", "Omar", "Zara"}

# TODO: union (|)
print(group_a | group_b)

# TODO: intersection (&)
print(group_a & group_b)
```

## Exercise 6 — ⭐⭐ List of dictionaries (the bridge!)
Yeh kal CSV ban jayega. `products` list di gayi hai.
1. Loop chala kar har product ka `name` aur `price` print karo
2. Total kitne products hain print karo (`len`)

```python
products = [
    {"name": "Pen",      "price": 50},
    {"name": "Notebook", "price": 120},
    {"name": "Bag",      "price": 800},
]

# TODO: loop and print "name - price" for each
for p in products:
    print(p["name"], ":" , p["price"])

# TODO: print total count
len(products)
```
